# Python Bloomberg API

Based on example provided by Bloomberg Open API, interface similar with Excel bdh function to Bloomberg is made. The output of bdh function is pandas dataframe with hierachical index.

## Bloomberg Open API Example

```python
import blpapi
from collections import defaultdict
from pandas import Series, DataFrame, Panel
from datetime import datetime, date, time
import pandas as pd
```

* Session start

```python
# Fill SessionOptions
sessionOptions = blpapi.SessionOptions()
sessionOptions.setServerHost('localhost')
sessionOptions.setServerPort(8194)

# Create a Session
session = blpapi.Session(sessionOptions)

# Start a Session
if not session.start():
    print "Failed to start session."

session.nextEvent()
```

* opening service

```python
# Open service to get historical data from
if not session.openService("//blp/refdata"):
    print "Failed to open //blp/refdata"
session.nextEvent()
# Obtain previously opened service
refDataService = session.getService("//blp/refdata")
session.nextEvent()
```

* request historical data

```python
# Create and fill the request for the historical data
request = refDataService.createRequest("HistoricalDataRequest")
request.getElement("securities").appendValue("KOSPI2 Index")
request.getElement("securities").appendValue("KOSPI Index")
request.getElement("fields").appendValue("PX_LAST")
request.getElement("fields").appendValue("OPEN")
request.set("periodicityAdjustment", "ACTUAL")
request.set("periodicitySelection", "DAILY")
request.set("startDate", "20060101")
request.set("endDate", "20061231")

print "Sending Request:", request
# Send the request
session.sendRequest(request)
```

* Receiving data

```python
field_to_request = ["PX Last", "OPEN"]
data = defaultdict(dict)
# Process received events
while(True):
    # We provide timeout to give the chance for Ctrl+C handling:
    ev = session.nextEvent(500)
    for msg in ev:
        ticker = msg.getElement('securityData').getElement('security').getValue()
        fieldData = msg.getElement('securityData').getElement('fieldData')
        for i in range(fieldData.numValues()) :
            for j in range(1,fieldData.getValue(i).numElements()) :
                data[(ticker, field_to_request[j-1])][fieldData.getValue(i).getElement(0).getValue()] = fieldData.getValue(i).getElement(j).getValue()

    if ev.eventType() == blpapi.Event.RESPONSE:
        # Response completly received, so we could exit
        break
```

```python
multi_test = DataFrame(data)
multi_test.columns = pd.MultiIndex.from_tuples(multi_test)
multi_test.head()
```

* Stop session

```python
# Stop the session
session.stop()
```

* Session start function

```python
def session_start(host='localhost', port=8194) :
    # Fill SessionOptions
    sessionOptions = blpapi.SessionOptions()
    sessionOptions.setServerHost('localhost')
    sessionOptions.setServerPort(8194)

    # Create a Session
    session = blpapi.Session(sessionOptions)

    # Start a Session
    if not session.start():
        print "Failed to start session."
    
    session.nextEvent()
    
    return session
```

* Opening Service for refData

```python
def service_refData(session):
    # Open service to get historical data from
    if not session.openService("//blp/refdata"):
        print "Failed to open //blp/refdata"
    
    session.nextEvent()
    
    # Obtain previously opened service
    refDataService = session.getService("//blp/refdata")
    
    session.nextEvent()
    
    return refDataService
```

* bdh(ticker, fld_list, start_date, end_date=date.today().strftime('%Y%m%d'), periodselection = 'DAILY') - return pandas multilevel columns

```python
def bdh(session, refDataService, ticker_list, fld_list, start_date, end_date=date.today().strftime('%Y%m%d'), periodselection = 'DAILY'):
    # Create and fill the request for the historical data
    request = refDataService.createRequest("HistoricalDataRequest")
    for t in ticker_list:
        request.getElement("securities").appendValue(t)
    for f in fld_list :
        request.getElement("fields").appendValue(f)
    request.set("periodicityAdjustment", "ACTUAL")
    request.set("periodicitySelection", periodselection)
    request.set("startDate", start_date)
    request.set("endDate", end_date)
    
    print "Sending Request:", request
    # Send the request
    session.sendRequest(request)
    # defaultdict - later convert to pandas
    data = defaultdict(dict)
    # Process received events
    while(True):
        # We provide timeout to give the chance for Ctrl+C handling:
        ev = session.nextEvent(500)
        for msg in ev:
            ticker = msg.getElement('securityData').getElement('security').getValue()
            fieldData = msg.getElement('securityData').getElement('fieldData')
            for i in range(fieldData.numValues()) :
                for j in range(1,fieldData.getValue(i).numElements()) :
                    data[(ticker, fld_list[j-1])][fieldData.getValue(i).getElement(0).getValue()] = fieldData.getValue(i).getElement(j).getValue()
    
        if ev.eventType() == blpapi.Event.RESPONSE:
            # Response completly received, so we could exit
            break
    data = DataFrame(data)
    data.columns = pd.MultiIndex.from_tuples(data, names=['ticker', 'field'])
    return data
```
